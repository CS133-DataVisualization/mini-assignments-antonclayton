```python
## Create a heatmap
# Your code here . . .
plt.figure(figsize=(10,6))
sns.heatmap(ign_scores, annot=True, cmap="coolwarm")
plt.title("IGN Scores by Genre and Platform")
plt.show()
```

## Problem 5 - (4 points)
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
# Problem 5
# Your code here . . .
pokemon_df = pd.read_csv("https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv")

type1_filter = ["Water", "Normal", "Grass", "Bug", "Psychic"]
filtered_pokemon_df = pokemon_df[pokemon_df["Type 1"].isin(type1_filter)]
filtered_pokemon_df.set_index("Name", inplace=True)
# filtered_pokemon_df

# colors for each type
color_map_type1 = {
    "Water": "blue",
    "Normal": "gray",
    "Grass" : "green",
    "Bug" : "darkgreen",
    "Psychic" : "pink"
}

mapped_color = filtered_pokemon_df["Type 1"].map(color_map_type1)

sns.clustermap(
    filtered_pokemon_df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', "Speed"]],
    cmap="coolwarm",
    row_colors=mapped_color,
    figsize=(12,12)
    )

plt.title("Clustermap of Pokemon by Type 1 with All Pokemon | Colors: Water=BLUE, Normal=GRAY, GRASS=green, BUG=darkgreen, Psychic=PINK")
# plt.xlabel("Pokemon Stat Categories")
plt.ylabel("Pokemon Stat Number")
plt.show()
```
